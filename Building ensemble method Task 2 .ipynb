```python
#Importing the dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
#Data collection and preprocessing
df = pd.read_csv('heart_disease.csv')
```

```python
df.head()
```

```python
#converting the categoorical data into numerical data
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df['Exercise Habits'] = encoder.fit_transform(df['Exercise Habits'])
df['Smoking'] = encoder.fit_transform(df['Smoking'])
df['Family Heart Disease'] = encoder.fit_transform(df['Family Heart Disease'])
df['Diabetes'] = encoder.fit_transform(df['Diabetes'])
df['Low HDL Cholesterol'] = encoder.fit_transform(df['Low HDL Cholesterol'])
df['High LDL Cholesterol'] = encoder.fit_transform(df['High LDL Cholesterol'])
df['Alcohol Consumption'] = encoder.fit_transform(df['Alcohol Consumption'])
df['High Blood Pressure'] = encoder.fit_transform(df['High Blood Pressure'])
df['Stress Level'] = encoder.fit_transform(df['Stress Level'])
df['Sugar Consumption'] = encoder.fit_transform(df['Sugar Consumption'])
df['Heart Disease Status'] = encoder.fit_transform(df['Heart Disease Status'])
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
# checking for missing value
df.isnull().sum()
```

```python
#filling the missing value with mean and mode
df['Age'].fillna(df['Age'].mean())
df['Blood Pressure'].fillna(df['Blood Pressure'].mean())
df['Cholesterol Level'].fillna(df['Cholesterol Level'].mean())
df['BMI'].fillna(df['BMI'].mean())
df['Sleep Hours'].fillna(df['Sleep Hours'].mean())
df['Triglyceride Level'].fillna(df['Triglyceride Level'].mean())
df['Fasting Blood Sugar'].fillna(df['Fasting Blood Sugar'].mean())
df['CRP Level'].fillna(df['CRP Level'].mean())
df['Homocysteine Level'].fillna(df['Homocysteine Level'].mean())
```

```python
df.isnull().sum()
```

```python
# spliting the feature and target value
X = df.drop(columns= ['Heart Disease Status'])

Y = df['Heart Disease Status']
```

```python
X

Y

print(X.shape, Y.shape)
```

```python
#Spliting the train and test data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2,stratify = Y, random_state =2)
```

```python
rf = RandomForestClassifier()
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
#Model Training
rf.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
#prediction
pred = rf.predict(X_test)
```

```python
print('Accuracy score:',accuracy_score(Y_test, pred))
```

```python
print("Accuracy:", accuracy_score(Y_test, pred))
print(classification_report(Y_test, pred))
```

```python
# Hyperparameters using GridSearchCV
#importing the gridsearchcv
from sklearn.model_selection import GridSearchCV
```

```python
#GridSEarchCV
#Gridsearchcv is use for determining the best parameter for our models
model = rf
print(rf.get_params())
#print(rf.get_params().keys())
```

```python
param_grid = {'n_estimators':[100, 200, 300],
              'max_depth':[5,10,None],
              'min_samples_split':[2,5,10,],}
             # 'min_samples_leaf':[1,2,4],
              #'max_features':['sqrt','log2',None]}
```

```python
grid = GridSearchCV(estimator = rf,
                   param_grid = param_grid,
                   cv=3)
```

```python
grid.fit(X_train,Y_train)
```

```python
print('Best Paramters', grid.best_params_ )
print('Best Score', grid.best_score_)
```
